# Data Preprocessing

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("../Data/data.csv")
X = dataset.iloc[:, :-1].values # taking everything except last col
y = dataset.iloc[:, -1].values
X, y
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(X[:, 1:3]) # selecting only numerical values
X[:, 1:3] = imputer.transform(X[:, 1:3]) # changing the only cols which have missing values
X
```

## Encoding categorical data

We encode categorical data or text data so that everything becomes machine-readable mathematically

### Encoding independent variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
X
```

### Encoding dependent variable

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
y
```

## Splitting the dataset into Training & Testing set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1) # seeding the dataset with '1'
X_train, X_test, y_train, y_test
```

## Features Scaling

### Standardization 

Algorithms assume normal/gaussian like data

- Examples:
     - Logistic regression
     - Linear regression
     - SVM
     - PCA
     - Neural networks

When data has outliners, use standardization 

### Normalization
 
Use this technique when bounded data is needed

- Examples:
    - Image processing
    - Deep Learning
    - KNN
    - K-Means clustering
    - Distance-based algorithms


Normalization prevents large-value features from dominating distance

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test = sc.transform(X_test[:, 3:])

X_train, X_test
```
