# Decision Tree Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../Data/salary_position.csv")
dataset
```

```python
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values

X, y
```

## Training the Decision Tree Regression Model on the dataset

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state = 0)
model.fit(X, y)
```

## Predicting a new result

```python
model.predict([[5.5]])
```

## Visualizing Decsion Tree Regression Results (Higher Resolution)

```python
X_grid = np.arange(np.min(X), np.max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, model.predict(X_grid), color = 'blue')
plt.xlabel("Position")
plt.ylabel("Salary")
```
