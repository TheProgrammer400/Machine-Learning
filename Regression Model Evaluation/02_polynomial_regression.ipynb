# Polynomial Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import The Dataset

```python
dataset = pd.read_csv("../Data/Data.csv")
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X, y
```

## Splitting The Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Traing the Polynomial Linear Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X_train)
model = LinearRegression()
model.fit(x_poly, y_train)
```

## Predicting the test results

```python
y_pred = model.predict(poly_reg.fit_transform(X_test))

for predicted, actual in zip(y_pred, y_test):
    print(f"Predicted: {predicted:.2f}       Actual: {actual:.2f}")
```

## Evaluating The Model Performance

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred) # found to be better than the multi linear regression model
```
