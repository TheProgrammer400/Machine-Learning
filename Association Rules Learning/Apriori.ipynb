# Apriori

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("../Data/Market_Basket_Optimisation.csv", header = None)
dataset
```

## Data-Preprocessing

```python
transactions = []

for i in range(len(dataset)):
    transactions.append([str(dataset.values[i, j]) for j in range(20)])
```

## Training the Apiori Model on the dataset

```python
from apyori import apriori

rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualizing the Results

### Displaying first result coming directly from the output of apriori function

```python
results = list(rules)
results
```

### Putting results into well organized `pandas` dataframe

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]

    return list(zip(lhs, rhs, supports, confidences, lifts))

resultDataframe = pd.DataFrame(inspect(results), columns = ['Left hand side', 'Right hand side', 'Support', 'Confidence', 'Lift'])
resultDataframe
```

## Display the sorted results

```python
resultDataframe = resultDataframe.sort_values(
    by='Lift',
    ascending=True
).reset_index(drop=True)

resultDataframe
```
